## Q: Which languages are the most people interested to learn over the next year?
For this, we can use the LanguageDesireNextYear column, with similar processing as the previous one.

```python
#col_series=dataframe.column_name
def split_col(col_series):
    new_df=col_series.to_frame()
    options=[]
    for idx,value in col_series[col_series.notnull()].iteritems():
        for option in value.split(';'):
            if not option in new_df.columns:
                options.append(option)
                new_df[option]=False
            new_df.at[idx,option]=True
    return new_df[options]
```

```python
lag=split_col(survey_df.LanguageWantToWorkWith)
lag
```

```python
future_popular_lang=lag.mean().sort_values(ascending=False)*100
future_popular_lang
```

```python
plt.title('most common languages which will be used')
matplotlib.rc('ytick',labelsize=10)
future_popular_lang=future_popular_lang.head(25)
sns.barplot(x=future_popular_lang,y=future_popular_lang.index,orient='h')
```

## Q: Which are the most loved languages, i.e., a high percentage of people who have used the language want to continue learning & using it over the next year?¶
While this question may seem tricky at first, it's straightforward to solve using Pandas array operations. Here's what we can do:

* Create a new data frame languages_loved_df that contains a True value for a language only if the corresponding values in languages_worked_df and languages_interested_df are both True
* Take the column-wise sum of languages_loved_df and divide it by the column-wise sum of languages_worked_df to get the percentage of respondents who "love" the language
* Sort the results in decreasing order and plot a horizontal bar graph

```python
ddf=lag&dm
```

```python
most_loved_lang=ddf.mean().sort_values(ascending=False)*100
most_loved_lang.head(25)
```

```python
plt.title('most loved languages for now and future')
matplotlib.rc('ytick',labelsize=10)
most_loved_lang=most_loved_lang.head(25)
sns.barplot(x=most_loved_lang,y=most_loved_lang.index,orient='h')
```

```python
mmf=dm&~lag
```

```python
most_dreaded_lang=mmf.mean().sort_values(ascending=False)*100
most_dreaded_lang.head(25)
```

```python
plt.title('most dreaded languages')
matplotlib.rc('ytick',labelsize=10)
most_dreaded_lang=most_dreaded_lang.head(25)
sns.barplot(x=most_dreaded_lang,y=most_dreaded_lang.index,orient='h')
```
