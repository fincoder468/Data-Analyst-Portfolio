### Q.5 Most Popular Cuisines served throughout the dataset?

```python
df
```

```python
# splitting cuisines
def split_col(cuisine):
    col_lis=[]
    new_df=cuisine.to_frame()
    for idx,value in cuisine[cuisine.notnull()].items():
        for val in value.split(';'):
            if not val in new_df.columns:
                new_df[val]=False
                col_lis.append(val)
            new_df.at[idx,val]=True
    return new_df[col_lis]
```

```python
new_df=split_col(df.cuisine)
new_df
```

```python
new_df.sum()
```

```python
def div(df):
    col=df.rating_count
    df['no._of_rating']=""
    for idx,value in col.items():
        if pd.isna(value):
            df.at[idx,'no._of_rating']=int(0)
        else:
            lis=list(value.split('+'))
            try:
                df.loc[idx,'no._of_rating']=int(lis[0])
            except:
                df.at[idx,'no._of_rating']=int(0)
```

```python
div(df)
```
